### tool 리스트로

```python
def remove_galho(text:str):
  return text[1:-1]

know2017['label_tool'] = know2017['label_tool'].map(remove_galho)
```

```python
def text_to_list(text:str):
  result = []
  for i in text.split(','):
    result.append(int(i))
  return result

know2017['label_tool'] = know2017['label_tool'].map(text_to_list)
```

```python

def list_truncate(tool:list):
  if len(tool) > 3:
    return tool[:3]
  elif len(tool) == 3:
    return tool
  elif len(tool) == 2:
    return tool+['0']
  elif len(tool) == 1:
    return tool+['0', '0']
  elif len(tool) == 0:
    return ['0', '0', '0']

know2017['label_tool'] = know2017['label_tool'].map(list_truncate)

def list_to_feature1(tool:list):
  return tool[0]

def list_to_feature2(tool:list):
  return tool[1]

def list_to_feature3(tool:list):
  return tool[2]


feature1 = know2017['label_tool'].map(list_to_feature1)
feature2 = know2017['label_tool'].map(list_to_feature2)
feature3 = know2017['label_tool'].map(list_to_feature3)

know2017['feature1'] = feature1
know2017['feature2'] = feature2
know2017['feature3'] = feature3

know2017[['feature1', 'feature2', 'feature3']]
```

```python

def list_truncate(tool:list):
  if len(tool) > 3:
    return tool[:3]
  elif len(tool) == 3:
    return tool
  elif len(tool) == 2:
    return tool+['0']
  elif len(tool) == 1:
    return tool+['0', '0']
  elif len(tool) == 0:
    return ['0', '0', '0']

know2017['label_tool'] = know2017['label_tool'].map(list_truncate)
def list_to_feature1(tool:list):
  return tool[0]

major1 = know2017['label_major'].map(list_to_feature1)

know2017['major'] = major1
```
